```python
%matplotlib inline
import numpy as np
from diogenes.read import (open_csv,
                           open_csv_url,
                           connect_sql)
from diogenes.utils import convert_to_sa
```

```python
data = open_csv_url(
            'http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv',  
            delimiter=';')
```

The default behavior expects a csv with the first row as a header.

In this data set, we've decided to classify according to the column which has the col_name 'quality'. 

So we assign y the values in the 'quality' column.  Next we convert this to a binary classification problem to accomdate the report generation.  

```python
labels = data['quality']
lables = labels < np.average(labels)
```
